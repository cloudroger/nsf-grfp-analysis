```python
import pandas as pd
import glob
```

```python
files = sorted(glob.glob('data/*'))
print(files[-1])
```

```python
df = pd.DataFrame()

for f in files: # build dataframe including award year
    dfn = pd.read_csv(f, sep='\t', engine='python')
    dfn['Year'] = int(f.split('/')[1].split('-')[0])
    df = pd.concat([df, dfn])
```
